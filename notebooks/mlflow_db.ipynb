```python
from pathlib import Path
import polars as pl
import numpy as np
from sklearn.metrics import (
    f1_score,
    balanced_accuracy_score,
    recall_score,
    log_loss,
)
from scipy.stats import entropy
```

```python
def evaluate_model(model_dir: Path, filename: str = "val_predictions.parquet") -> dict:
    file_path = model_dir / filename
    
    df = pl.read_parquet(file_path)

    # Ensure correct types
    y_true = df["true_label"].to_numpy()
    y_pred = df["pred_label"].to_numpy()

    prob_cols = [
        "prob_angry",
        "prob_fear",
        "prob_happy",
        "prob_sad",
        "prob_surprise",
    ]
    probs = df.select(prob_cols).to_numpy()

    labels = ["angry", "fear", "happy", "sad", "surprise"]
    label_mappings = {'angry': 0, 'fear': 1, 'happy': 2, 'sad': 3, 'surprise': 4}
    # ---- Metrics ----
    macro_f1 = f1_score(y_true, y_pred, average="macro")
    weighted_f1 = f1_score(y_true, y_pred, average="weighted")
    bal_acc = balanced_accuracy_score(y_true, y_pred)

    logloss = log_loss(y_true, probs)

    per_class_recall = recall_score(
        y_true,
        y_pred,
        average=None,
        # labels=labels,
    )

    # support = df.groupby("true_label").count().sort("true_label")
    # support_dict = dict(zip(support["true_label"], support["count"]))
    
    
    # ---- Confidence diagnostics ----
    mean_confidence = probs.max(axis=1).mean()
    mean_entropy = entropy(probs.T).mean()

    # ---- Validation size ----
    val_size = len(df)

    # ---- Build output row ----
    result = {
        "model_name": model_dir.name,
        "val_size": val_size,
        "macro_f1": macro_f1,
        "weighted_f1": weighted_f1,
        "balanced_accuracy": bal_acc,
        "log_loss": logloss,
        "mean_confidence": mean_confidence,
        "mean_entropy": mean_entropy,
    }

    # Add per-class recall
    for label, recall_val in zip(labels, per_class_recall):
        result[f"recall_{label}"] = recall_val

    # Add per-class support
    for label in labels:
        result[f"support_{label}"] = df.select(
            (pl.col("emotion") == label).sum()).item()

    return result


# --------------------------------------------------
# Batch evaluation function
# --------------------------------------------------

def evaluate_multiple_models(model_dirs, filename="val_predictions.parquet"):
    results = []
    
    for model_dir in model_dirs:
        model_dir = Path(model_dir)
        results.append(evaluate_model(model_dir, filename))
    
    return pl.DataFrame(results)
```

```python
results_df = evaluate_multiple_models(my_models)

results_df = results_df.sort("macro_f1", descending=True)
```
